<h2> Apa itu machine learning? </h2>

- machine learning merupakan cabang ilmu pengentahuan dimana membuat komputer dapat menyelsaikan masalah dengan cara belajar dari data 

<h2> Apa itu data feature dan data target </h2>

- data feature merupakan data-data yang mewakili dari label itu sendiri

- data target merupakan data label yang digunakan untuk melakukan prediksi

<h2> Apa Perbedaan Supervised Learning, dan Unsupervised Learning </h2>

- supervised learning merupakan metode pembelajaran dimana dataset terdapat feature variable dan juga target variable (label)

- unsupervised learning merupakan metode pembelajaran dimana dataset terdapat feature variable saja tanpa adanya target variable (label)


<h2> Apa Jenis2 yang ada di dalam Supervised Learning? Jelaskan Perbedaannya! </h2>

- Regressi: sistem melakukan prediksi dimana output dari proram merupakan nilai kontinu

- Klasifikasi: sistem melakukan prediksi dimana output dari program merupakan kategori

<h2> Apa perbedaan Hyperparameter dan Parameter </h2>

- Hypyerparameter merupakan parameter yang  didefinisikan diawal pembuatan machine learning

- parameter merupakan merupakan parameter yang didapatkan diakhir proses training

<h1>Quiz 2 : Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian applikasikan untuk membuat model bagi data (variable x, y) di bawah ini. kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

Contoh Jawaban Hasil Pembuatan Model dengan Algoritma DecisionTreeRegressor (Jawaban tidak haru sama) :
![title](img/day1.png)

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as  np
X = x.reshape(-1,1)
Y = y.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X,Y)

lr = LinearRegression(normalize=True)
lr.fit(X_train,y_train)
```

```python
predict = lr.predict(x_new.reshape(-1,1))

plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=[15,10])
ax.scatter(X_train, y_train, c='b', label='Data Asli')
ax.scatter(X_test, lr.predict(X_test), c='green', label='Prediksi terhadap data training')
ax.plot(x_new[:20], predict, c='lightgreen', label='Prediksi data baru')
plt.legend()
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

print(f'Akurasi model sebesar\t\t: {lr.score(X_test,y_test)}')
print(f'besar dari RMSe adalah sebesar\t: {sqrt(mean_squared_error(y_train, lr.predict(X_train)))}')
```
